```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
d = 30
rho_all = [100, 500, 1000, 1500, 2000, 2500]
n = 50  # Ambient space dimensionality
L = 100  # Number of subspaces
N_all = []
for rho in rho_all:
    N_all.append(rho * L * d)
print(N_all)
```

```python
def plot_sdp_rate_hist(sdp_rates, rho):
    # 100 bins from 0 to 1
    bins = np.linspace(0, 1, 101)  # 101 edges will create 100 bins
    plt.hist(sdp_rates, bins=bins, alpha=0.75, edgecolor='black')
    plt.title('Histogram of SDP rates for rho={}'.format(rho))
    plt.xlabel('SDP rate')
    plt.ylabel('Frequency')
    plt.show()
```

```python
for rho in rho_all:
    sdp_rates = np.load('/home/local/ASUAD/ywan1053/basis/SDP_results/sdp_rho_{}.npy'.format(rho))
    plot_sdp_rate_hist(sdp_rates, rho)
```
