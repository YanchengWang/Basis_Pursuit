```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

rho=100
n=50 
L=100
dimensions = [10, 20, 30, 40]

dir = '/home/local/ASUAD/ywan1053/basis/epsilon_0'

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'figure.autolayout': True})
# chaneg the font size
plt.rcParams.update({'font.size': 15})
def plot_historgam(support_size, d, max_size=50):
    # plot a histogram of the support with border
    plt.hist(support_size, bins=range(0, max_size+1, 1), alpha=0.6, edgecolor='black')
    # plt.title('Histogram of support sizes for d={}'.format(d))
    plt.xlabel('Support Size')
    plt.ylabel('Frequency')
    plt.savefig(dir + '/histograms/histogram_d_{}.pdf'.format(d))
    plt.show()
    plt.close()

for d in dimensions:
    support_size = np.load(dir + '/support_sizes_d_{}.npy'.format(d))
    plot_historgam(support_size, d)
```
